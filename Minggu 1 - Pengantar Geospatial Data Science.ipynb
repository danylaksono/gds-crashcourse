### *Latihan 1: Menerapkan Fungsi Dasar Python*

Latihan ini bertujuan untuk memberikan pemahaman mengenai bagaimana fungsi-fungsi dasar dari Python dapat digunakan dalam pengolahan data. *List* merupakan salah satu jenis *container* yang paling sering dijumpai pada pengolahan data untuk data science, sehingga pada bagian ini akan diberikan contoh pengolahan data dengan list.

````
## Membuat list sederhana dan melakukan operasi dasar pada list

# Membuat list
xs = [3, 1, 2]    
print(xs, xs[2])  

# menggunakan index untuk mengakses List
print(xs[-1])     
xs[2] = 'foo'     
print(xs)

# menambahkan elemen baru pada list
xs.append('bar')  
print(xs)

# menghapus elemen terakhir dari isi list
x = xs.pop()      
print(x, xs)      
````

Perhatikan bagaimana elemen dari List digunakan pada contoh di atas. Selanjutnya, beberapa teknik pada List yang sering digunakan adalah sebagai berikut:

````
# teknik slicing

nums = list(range(5))     
print(nums)               

# bagaimana keluaran dari tiap slicing berikut?
print(nums[2:4])          
print(nums[2:])           
print(nums[:2])           
print(nums[:])            
print(nums[:-1])          
nums[2:4] = [8, 9]        
print(nums)  
````

Dengan menggunakan Slicing, pengolahan pada sebuah List akan menjadi sangat mudah: cukup gunakan indeks yang diinginkan untuk melakukan manipulasi data. Teknik inilah yang tadi kita gunakan pada contoh untuk hitungan bilangan Fibonacci

````
# menampilkan isi dari list
animals = ['kucing', 'cacing', 'anjing']
for animal in animals:
    print(animal)
````

Cara di atas digunakan untuk menampilkan isi dari sebuah list. Ini adalah teknik yang juga digunakan pada bahasa pemrograman lain: kita dapat menggunakan iterasi untuk menampilkan isi dari sebuah list. Seringkali, kita juga perlu melakukan fungsi tertentu pada tiap elemen list tersebut, sebagai contoh:

````
nums = [0, 1, 2, 3, 4]
squares = []
for x in nums:
    squares.append(x ** 2)
print(squares)
````

Baris perintah di atas mengisi array `squares` dengan kuadrat dari tiap angka pada array `nums`. Bagaimana dengan ini?

````
nums = [0, 1, 2, 3, 4]
squares = [x ** 2 for x in nums]
print(squares)
````

Teknik di atas disebut dengan nama **List Comprehension**, dan akan banyak kita jumpai pada pengolahan data pada Data Science. Teknik tersebut juga dapat diterapkan pada Dictionary, dan disebut sebagai **Dictionary Comprehension**

**LATIHAN**

Sebagai **Latihan**, buat sebuah program Python untuk menampilkan angka kuadrat dari bilangan 1 sampai 10 seperti list di atas, tetapi hanya untuk bilangan genap saja

### *Latihan 3: Menggunakan Modul Python untuk Geospatial Data Science*

Sebagaimana telah disebutkan, Python memiliki modul yang sangat kaya untuk keperluan analisis data. Pada contoh berikut kita akan mengolah data spasial sederhana dan menampilkan data tersebut pada IPython yang akan kita gunakan berikut ini.



````
import numpy
import scipy.stats
import pandas
import matplotlib.pyplot as plt
plt.style.use("bmh")
````

````
# membaca data gempabumi dari USGS
data = pandas.read_csv("https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/2.5_month.csv")
data.head()
````

````
# berapa jumlah data yang diolah?
len(data)
````

````
# urutkan berdasarkan waktu terjadinya
data.time = pandas.to_datetime(data.time)
data.sort_values("time", inplace=True)
````

````

# coba kita plot data gempa ini
fig = plt.figure()
fig.set_size_inches(20, 4)
plt.plot(data.time, data.mag, ".");
plt.ylabel("Magnitude");

````

Apa yang dapat Anda simpulkan dari plot di atas?

Tidak ada!. Tidak ada pola tertentu yang dapat kita amati. Untuk itulah, proses pengolahan data pada Data Science perlu dilakukan secara berulang, yaitu agar kita dapat memahami pola awal dari data yang kita miliki. Untuk itu kita dapat menggunakan tampilan lain agar lebih jelas:

````
# tampilkan dalam bentuk histogram
data.mag.hist(density=True, alpha=0.5, bins=30)
plt.xlabel("Earthquake magnitude")
plt.title("Histogram of earthquake magnitudes");

````

Dari data di atas kita dapat melihat satu pola tertentu: Gempa dengan Magnitudo lebih dari 5 dapat dikatakan cukup jarang terjadi. Kita juga dapat melakukan ekstraksi informasi tertentu, misalnya:

````
#Berapa jumlah gempa dalam satu hari?
duration = data.time.max() - data.time.min()
density = len(data) / float(duration.days)
density  
````

````
# gunakan numpy untuk membuat distribusi exponensial dan matplotlib untuk plotting
interarrival = data.time.diff().dropna().apply(lambda x: x / numpy.timedelta64(1, "D"))
support = numpy.linspace(interarrival.min(), interarrival.max(), 100)
interarrival.hist(density=True, alpha=0.5, bins=30)
plt.plot(support, scipy.stats.expon(scale=1/density).pdf(support), lw=2)
plt.title("Jeda antara 5> Gempa bumi", weight="bold")
plt.xlabel("hari");
````

````
# probability plot

shape, loc = scipy.stats.expon.fit(interarrival)
scipy.stats.probplot(interarrival, 
                     dist="expon", sparams=(shape, loc), 
                     plot=plt.figure().add_subplot(111))
plt.title("Exponential probability plot pada waktu antar gempa");

````

````
# Autokorelasi
plt.acorr(interarrival, maxlags=200)
plt.title("Autocorrelation of earthquake timeseries data");
````

### *Latihan 4: Visualisasi Data menggunakan Folium*

Salah satu metode dalam melakukan *data wrangling* adalah dengan menampilkan bagaimana persebaran data tersebut. Pada kasus data spasial, tentunya hal ini menjadi sangat relevan. Berikut adalah latihan untuk menampilkan data dengan menggunakan library Folium.


````
# Instalasi library jika belum ada pada Google Colab
!pip install folium
````

````
import folium
import pandas as pd

# mengambil data gempabumi dari API USGS
usgs_url = 'https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/2.5_month.csv'
df = pd.read_csv(usgs_url)

df.info() 
````

````
# melihat data
df.head()
````

````
# memanggil folium
m = folium.Map(location=(0, 0), zoom_start=2)

# save our map to an interactive html file
m
````

````
for i in range(len(df)):
    folium.Circle(
        location=[df.iloc[i]['latitude'], df.iloc[i]['longitude']],
        radius=10,
    ).add_to(m)
m
````

````
# untuk setiap titik yang ada pada data, buat lingkaran dengan simbologi folium
for i in range(len(df)):
    folium.Circle(
        location=[df.iloc[i]['latitude'], df.iloc[i]['longitude']],
        radius=df.iloc[i]['mag'] * 50000,
        fill_opacity=0.1,
        
    ).add_to(m)
m
````

````
# sama seperti sebelumnya, tetapi dengan simbologi yang lebih keren
for i in range(len(df)):
    if df.iloc[i]['mag'] > 3:
        folium.Circle(
            location=[df.iloc[i]['latitude'], df.iloc[i]['longitude']],
            radius=df.iloc[i]['mag'] * 70000,
            weight=1,  # thickness of the border
            color='red',  # this is the color of the border
            opacity=0.3,  # this is the alpha for the border
            fill_color='red',  # fill is inside the circle
            fill_opacity=0.1,  # we will make that less opaque so we can see layers

        ).add_to(m)
m
````

````
url = 'https://raw.githubusercontent.com/fraxen/tectonicplates/master/GeoJSON/PB2002_boundaries.json'

folium.GeoJson(
    url,
    name='geojson'
).add_to(m)


folium.LayerControl().add_to(m)

m
````
